# This notebook takes msid_mbid_filtered_mapping_file.csv created in Data pre-processing-2.ipynb and goes on doing mappings between .csv tables only with dictionaries, without additional .csv files like in the previous approach
The takeaway of this is that dictionaries are "extended" sets, and I did not really needed sets for this assignment.
I used sets to check whether a particular element (row of a .csv file) is contained or not in a group of other elements. I can easily retrieve a set from a dictionary's keys or values

```python
import json # extract data from the monthly json lines files
import time # measuring time for .csv files for loop iterations
# import numpy
'''
Data serialization/de-serialization
Once I compute the extraction of some data of interest
out of a larger data-set, I wish not to repeat that operation
over and over each time I run this notebook.
I tested it and de-serialization is pretty fast
(4000000 3 keys dictionaries de-serialized in 4 seconds)
'''
import pickle # for each recording in Listenbrainz 2022 month
import csv # for reading -and creating a smaller "version" of- listenbrainz_msid_mapping.csv
```

```python
messyBrainzID_to_MusicBrainzID_mappingsDict = dict()
with open('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/listenbrainz_msid_mapping_filtered.csv') as listenbrainz_msid_mapping_filtered_file:
  csvReader = csv.DictReader(listenbrainz_msid_mapping_filtered_file, ['recording_msid','recording_mbid'])
  start = int(time.monotonic())
  for csvLine in csvReader:
    # first column is MessyBrainz id
    messyBrainzID_to_MusicBrainzID_mappingsDict[csvLine['recording_msid']] = csvLine['recording_mbid']
  end = int(time.monotonic()) # end of csvReader iteration
  print(f'Finished creating a dictionary of MessyBrainzID (keys) <-> MusicBrainzID (values) mappings')

# pickle dump
messyBrainzID_to_MusicBrainzID_mappingsDict_file_path = str('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/Dictionary_of_MessyBrainzID_to_MusicBrainzID_mappings_filtered.p')
pickle.dump(messyBrainzID_to_MusicBrainzID_mappingsDict, open(messyBrainzID_to_MusicBrainzID_mappingsDict_file_path, "wb" ) )
# FREE MEMORY UP
# del messyBrainzID_to_MusicBrainzID_mappingsDict
```

# Transform the UserID_to_MessyBrainzID_mappings_filtered.csv file made in the other notebook, into UserID_to_ArtistID_mappings_filtered.csv (create a new .csv file with user_id <-> artist_names mappings)

```python
userID_set = set()

with open('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/UserID_ArtistMBID_ArtistName_mappings_filtered.csv', "w") as userID_to_ArtistName_mappings_filtered_file:
  fieldnames = ['user_id', 'artist_mbid', 'artist_name']
  csvWriter = csv.DictWriter(userID_to_ArtistName_mappings_filtered_file, fieldnames=fieldnames)
  csvWriter.writeheader()
  with open('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/UserID_to_MessyBrainzID_mappings_filtered.csv') as userID_to_MessyBrainzID_mappings_filtered_file:
    csvReader = csv.DictReader(userID_to_MessyBrainzID_mappings_filtered_file)
    start = int(time.monotonic())
    for csvLine in csvReader:
      if csvLine['recording_msid'] in messyBrainzID_to_MusicBrainzID_mappingsDict:
        if messyBrainzID_to_MusicBrainzID_mappingsDict[csvLine['recording_msid']] in musicBrainzID_to_ArtistID_mappingsDict:
          if musicBrainzID_to_ArtistID_mappingsDict[messyBrainzID_to_MusicBrainzID_mappingsDict[csvLine['recording_msid']]] in artistID_to_ArtistName_mappingsDict:
            artist_name = artistID_to_ArtistName_mappingsDict[ musicBrainzID_to_ArtistID_mappingsDict[ messyBrainzID_to_MusicBrainzID_mappingsDict[csvLine['recording_msid']]]]
            # print(artist_name)
            csvWriter.writerow({'user_id': csvLine['user_id'], 'artist_mbid':musicBrainzID_to_ArtistID_mappingsDict[messyBrainzID_to_MusicBrainzID_mappingsDict[csvLine['recording_msid']]], 'artist_name': artist_name})
            userID_set.add(csvLine['user_id'])
    end = int(time.monotonic()) # end of csvReader iteration
    print(f'Finished creating user_id <-> ArtistName mappings .csv file, it took {end - start} seconds.')

# pickle dump
userID_set_file_path = str('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/Set_Of_UserIDs_filtered.p')
pickle.dump(userID_set, open(userID_set_file_path, "wb" ) )
# FREE MEMORY UP
# del userID_set # this is no longer needed
```

# Create a dictionary with artist name as keys, and list of dictionaries as value. Each nested dictionary inside the list of dictionaries, contains a user_id key and a listenings count key

This data structure will be used to create the final .csv file

```python
!head -50 '/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/UserID_ArtistMBID_ArtistName_mappings_filtered.csv'
```

```python
artistName_to_ListCounts_forUserID_nestedDict = dict()

with open('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/UserID_ArtistMBID_ArtistName_mappings_filtered.csv') as userID_ArtistMBID_ArtistName_mappings_filtered_file:
  csvReader = csv.DictReader(userID_ArtistMBID_ArtistName_mappings_filtered_file)
  start = int(time.monotonic())
  for csvLine in csvReader:
    # if an artist is already in the dict
    if csvLine['artist_name'] in artistName_to_ListCounts_forUserID_nestedDict.keys():
      # iterate through list of dictionaries for this artist
      for dict_Num in range(len(artistName_to_ListCounts_forUserID_nestedDict[csvLine['artist_name']])):
        # if this user is already in this artist's dictionary, increment count
        if csvLine['user_id'] in artistName_to_ListCounts_forUserID_nestedDict[csvLine['artist_name']][dict_Num].keys():
          artistName_to_ListCounts_forUserID_nestedDict[csvLine['artist_name']][dict_Num]['listenings_count'] += 1
        else:
        # if this user is not already in this artist's dictionary, add user dictionaries to this user's list of dictionaries and set count to 1
          artistName_to_ListCounts_forUserID_nestedDict[csvLine['artist_name']].append({'user_id': csvLine['user_id'], 'listenings_count': 1})
    # if an artist is not already in the dict, create a new list of dictionaries with only 1 dictionary for this user
    else:
      artistName_to_ListCounts_forUserID_nestedDict['artist_name'] = csvLine['artist_name']
      artistName_to_ListCounts_forUserID_nestedDict[csvLine['artist_name']] = [{'user_id': csvLine['user_id'], 'listenings_count': 1}]
    # print(artistName_to_ListCounts_forUserID_nestedDict)

# print(artistName_to_ListCounts_forUserID_nestedDict)

# pickle dump
artistName_to_ListCounts_forUserID_nestedDict_file_path = str('/content/gdrive/MyDrive/Colab Notebooks/Audio and Music Processing Lab/Audio and Music processing lab - assignment/Task 1/Data pre-processing-3/NestedDictionary_of_ArtistName_to_ListeningsCounts_forUserID.p')
pickle.dump(artistName_to_ListCounts_forUserID_nestedDict, open(artistName_to_ListCounts_forUserID_nestedDict_file_path, "wb" ) )
# FREE MEMORY UP
# del userID_toArtistID_Name_and_numList_dict # this is no longer needed
```
